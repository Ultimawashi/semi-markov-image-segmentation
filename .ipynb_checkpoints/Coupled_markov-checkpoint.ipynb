```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import multivariate_normal
from math import sqrt, pi,exp,log2
from random import random
from hilbertcurve.hilbertcurve import HilbertCurve
import sys
import itertools

np.seterr(all='raise')

def norm_pdf(Y,mu,sig):
    return(np.exp((-(Y-mu)**2)/(2*sig**2))/(sqrt(2*pi)*sig))
def multi_norm_pdf(Y,mu,sig):
    return(np.exp(-0.5*(Y-mu)@np.linalg.inv(sig)@(Y-mu))/((2*pi)**mu.shape[0]*sqrt(abs(np.linalg.det(sig)))))
```

```python
class Markov_Chain:
    
    def __init__(self,chain=np.array([]),categories=np.array([])):
        self.chain=chain
        self.n=self.chain.shape[0]
        self.categories=categories
        self.nb_categories=categories.shape[0]
        self.index_categories=dict()
        for i in range(self.nb_categories):
            self.index_categories[self.categories[i]]=i
            
    # Utilities
        
    def get_category(self):
        categories=np.unique(self.chain)
        self.categories=categories
        return(categories)
    
    def get_category_info(self):
        self.index_categories=dict()
        self.nb_categories=self.categories.shape[0]
        for i in range(self.nb_categories):
            self.index_categories[self.categories[i]]=i
        
    def init_from_image(self,image,categories=np.array([])):
        mat=np.asarray(image,)
        n=len(mat)
        chaine=np.zeros(n*n,dtype='int64')
        hilbert_curve=HilbertCurve(int(log2(n)),2)
        for i in range(n):
            for j in range(n):
                chaine[hilbert_curve.distance_from_coordinates([i,j])]=mat[i][j]
        self.chain=chaine
        self.n=chaine.shape[0]
        if(categories.size>0):
            self.categories=categories
        else:
            self.get_category()
        self.get_category_info()
        
    def get_image(self):
        n=len(self.chain)
        m=int(sqrt(n))
        mat=np.zeros((m,m))
        hilbert_curve=HilbertCurve(int(log2(n)),2)
        for i in range(n):
            coor=hilbert_curve.coordinates_from_distance(i)
            mat[coor[0]][coor[1]]=self.chain[i]
        return (mat)
        
    def plot_image(self,title=None,show=True):
        plt.title(title)
        plt.imshow(self.get_image(),cmap='Greys_r')
        if show:
            plt.show()
            
    def error_rate(self,B):
        mean=np.mean(np.abs(self.chain-B)/255)
        return(min(mean,1-mean)) 
      
    # Different noises
    
    def gaussian_noise(self,mean,std):        
        gauss=[]
        for i in range(self.nb_categories):
            gauss+=[np.random.normal(mean[i],std[i],self.n)]
        bruit=np.zeros((self.nb_categories,self.n))
        for i in range(self.nb_categories):
            bruit[i]=np.where(self.chain==self.categories[i],gauss[i],0)
        return(CMC(bruit.sum(axis=0),self.categories))
    
    def gaussian_noise_couple(self,mean,std):
        chaine_bruite=np.zeros(self.n)
        chaine_bruite[0]=np.random.normal(mean[self.index_categories[self.chain[0]]][self.index_categories[self.chain[1]]][0],std[self.index_categories[self.chain[0]]][self.index_categories[self.chain[1]]][0][0])
        for i in range(1,self.n):
            mean_i=mean[self.index_categories[self.chain[i-1]]][self.index_categories[self.chain[i]]]
            std_i=std[self.index_categories[self.chain[i-1]]][self.index_categories[self.chain[i]]]
            chaine_bruite[i]=np.random.normal(mean_i[1]+std_i[1][0]/std_i[0][0]*(chaine_bruite[i-1]-mean_i[0]),sqrt(std_i[1][1]-std_i[0][1]*std_i[1][0]/std_i[0][0]))            
        return(CMCouple(chaine_bruite,self.categories))
    
    # Compute probas
    
    def x_transtion_true(self):
    
        X=self.chain
        transitions=np.zeros((self.nb_categories,self.nb_categories))
        x_list=X.tolist()
        for (i,j) in zip(x_list,x_list[1:]):
            transitions[self.index_categories[int(i)]][self.index_categories[int(j)]]+=1
        transitions=(transitions.T/transitions.sum(axis=1)).T
        return(transitions)
    
    def x_joint_true(self):
        
        X=self.chain
        transitions=np.zeros((self.nb_categories,self.nb_categories))
        x_list=X.tolist()
        for (i,j) in zip(x_list,x_list[1:]):
            transitions[self.index_categories[int(i)]][self.index_categories[int(j)]]+=1
        return(transitions)
```

```python
class CMC(Markov_Chain):
    def __init__(self,chain=np.array([]),categories=np.array([])):
        
        super().__init__(chain,categories)

    def gaussian_assumption(self,mean,std):
        
        gaussians=np.zeros((self.nb_categories,self.n))
        for i in range(self.nb_categories):
            gaussians[i]=norm_pdf(self.chain,mean[i],std[i])
        return(gaussians)

    def forward(self,gaussians,n,transitions,p):
        
        alpha=np.zeros((self.nb_categories,n))
        alpha[:,0]=p*gaussians[:,0]
        alpha[:,0]=alpha[:,0]/(alpha[:,0].sum())
        for i in range(1,n):
            alpha[:,i]=gaussians[:,i]*np.transpose(alpha[:,i-1])@transitions
            alpha[:,i]=alpha[:,i]/(alpha[:,i].sum())
        return alpha
    
    def backward(self,gaussians,n,transitions):
        beta=np.zeros((self.nb_categories,n))
        beta[:,0]=np.ones(self.nb_categories)
        beta[:,0]=beta[:,0]/(beta[:,0].sum())
        for i in range(1,n):
            beta[:,i]=gaussians[:,i]*np.transpose(beta[:,i-1])@transitions
            beta[:,i]=beta[:,i]/(beta[:,i].sum())
        return beta
    
    def baum_welch(self,p0,transitions0,mean0,std0,nb_iter):
        
        p,transitions,mean,std=p0,transitions0,mean0,std0
        t=self.n
        Y=self.chain
        
        for _ in range(nb_iter):
            
            gaussians=self.gaussian_assumption(mean,std)
            alpha=self.forward(gaussians,t-1,transitions,p)
            beta=self.backward(gaussians,t,transitions)
            
            gamma=alpha[np.newaxis,:,:]*gaussians[:,np.newaxis,1:]*beta[:,np.newaxis,1:]*transitions[:,:,np.newaxis]
            gamma=gamma/(gamma.sum(axis=0).sum(axis=0))
            psi=gamma.sum(axis=0)
            
            transitions=(gamma.sum(axis=2))/(psi.sum(axis=1))
            mean=((psi*Y[np.newaxis,:-1]).sum(axis=1))/(psi.sum(axis=1))
            std=np.sqrt(((psi*(Y[np.newaxis,:-1]-mean[:,np.newaxis])**2).sum(axis=1))/(psi.sum(axis=1)))
            p=psi[:,0]
                       
        return(p,transitions,mean,std)
    
    def MPM(self,transitions0,p0,mean0,std0,nb_iter_bw=10):
        
        #Parameter Estimation
        p,transitions,mean,std=self.baum_welch(p0,transitions0,mean0,std0,nb_iter_bw)        
        # Prediction
        gaussians=self.gaussian_assumption(mean,std)
        alpha=self.forward(gaussians,self.n,transitions,p)
        beta=self.backward(gaussians,self.n,transitions)
        X_apost=np.zeros(self.n)
        p=alpha*beta
        categories_pmax=np.argmax(p,axis=0)
        
        return Markov_Chain(self.categories[categories_pmax],self.categories)
```

```python
class CMCouple(Markov_Chain):
    def __init__(self,chain=np.array([]),categories=np.array([])):
        
        super().__init__(chain,categories)

    

    def compute_probas_chaine(self,mean,std,x_joint):
        
        
        p_z_1=np.zeros((self.nb_categories,self.nb_categories))
        
        categories_combi=list(itertools.chain.from_iterable([list(zip(self.categories.tolist(),x)) for x in itertools.permutations(self.categories.tolist())]))
        for i,j in categories_combi:
            p_z_1[self.index_categories[i]][self.index_categories[j]]=norm_pdf(self.chain[0],mean[self.index_categories[i]][self.index_categories[j]][0],std[self.index_categories[i]][self.index_categories[j]][0][0])
        p_z_1=(p_z_1*x_joint).sum(axis=1)
        
        transi=np.zeros((self.nb_categories,self.nb_categories,self.n-1))

        for k in range(0,self.n-1):
            marginale=np.zeros((self.nb_categories,self.nb_categories))
            for i,j in categories_combi:
                marginale[self.index_categories[i]][self.index_categories[j]]=norm_pdf(self.chain[k],mean[self.index_categories[i]][self.index_categories[j]][0],std[self.index_categories[i]][self.index_categories[j]][0][0])
                transi[self.index_categories[i]][self.index_categories[j]][k]=multi_norm_pdf(self.chain[k:k+2],mean[self.index_categories[i]][self.index_categories[j]],std[self.index_categories[i]][self.index_categories[j]])
            marginale=(marginale*x_joint).sum(axis=1)
            transi[:,:,k]=(transi[:,:,k]*x_joint)/marginale
            
        return(p_z_1,transi)

    def forward(self,n,transitions,p_z_1):
        alpha=np.zeros((self.nb_categories,n))
        alpha[:,0]=p_z_1
        alpha[:,0]=alpha[:,0]/(alpha[:,0].sum())
        for i in range(1,n):   
            alpha[:,i]=np.transpose(alpha[:,i-1])@transitions[:,:,i-1]
            alpha[:,i]=alpha[:,i]/(alpha[:,i].sum())
        return alpha
    
    def backward(self,n,transitions):
        beta=np.zeros((self.nb_categories,n))
        beta[:,0]=np.ones(self.nb_categories)
        beta[:,0]=beta[:,0]/(beta[:,0].sum())
        for i in range(1,n):
            beta[:,i]=np.transpose(beta[:,i-1])@transitions[:,:,i-1]
            beta[:,i]=beta[:,i]/(beta[:,i].sum())
        return beta
    
    def baum_welch(self,p0,transitions0,mean0,std0,nb_iter):
        
        p,transitions,mean,std=p0,transitions0,mean0,std0
        t=self.n
        Y=self.chain
        
        for _ in range(nb_iter):
            
            gaussians=self.gaussian_assumption(mean,std)
            alpha=self.forward(gaussians,t-1,transitions,p)
            beta=self.backward(gaussians,t,transitions)
            
            gamma=alpha[np.newaxis,:,:]*gaussians[:,np.newaxis,1:]*beta[:,np.newaxis,1:]*transitions[:,:,np.newaxis]
            gamma=gamma/(gamma.sum(axis=0).sum(axis=0))
            psi=gamma.sum(axis=0)
            
            transitions=(gamma.sum(axis=2))/(psi.sum(axis=1))
            mean=((psi*Y[np.newaxis,:-1]).sum(axis=1))/(psi.sum(axis=1))
            std=np.sqrt(((psi*(Y[np.newaxis,:-1]-mean[:,np.newaxis])**2).sum(axis=1))/(psi.sum(axis=1)))
            p=psi[:,0]
                       
        return(p,transitions,mean,std)
    
    def MPM(self,transitions,p_z_1,nb_iter_bw=10):
        
        #Parameter Estimation
#         p,transitions,mean,std=self.baum_welch(p0,transitions0,mean0,std0,nb_iter_bw)
#         print(p,transitions,mean,std)
        
        # Prediction
        alpha=self.forward(self.n,transitions,p_z_1)
        beta=self.backward(self.n,transitions)
        X_apost=np.zeros(self.n)
        p=alpha*beta
        categories_pmax=np.argmax(p,axis=0)
        
        return Markov_Chain(self.categories[categories_pmax],self.categories)
```

```python
image_chain = Markov_Chain()
image_chain.init_from_image(np.array(Image.open('./images/beee2.bmp')))
mean=np.array([[[1,3],[2,4]],[[5,3],[1,1]]])
std=np.array([[[[1,0.5],[0.5,3]],[[1,0.5],[0.5,3]]],[[[1,0.5],[0.5,3]],[[1,0.5],[0.5,3]]]])
image_bruite=image_chain.gaussian_noise_couple(mean,std)
image_bruite.plot_image()
x_joint=image_chain.x_joint_true()
p_z_1,transitions=image_bruite.compute_probas_chaine(mean,std,x_joint)
image_segmente=image_bruite.MPM(transitions,p_z_1,10)
image_segmente.plot_image(title='error_rate = '+str(image_chain.error_rate(image_segmente)))
```
